```python
inc = 0
val = 0
val_inc = 0
all_tweets = []
tweet_id = []
tweet_created_at = []
followers_count = []
user_screen_name = []
retweet_id = []
retweeted_status = []
language = []
utc_offset = []
retweet_created_at = {}
tweet_text = {}
# filename = 'tweet_data/voterfraud_2018-11-06_09_28_37.json'
# filename = '../caravan/csv/caravan_main.csv'
filename = 'tweet_data/nerd.json'
print("Start:" + str(datetime.datetime.now()))

with open(filename, 'r') as f:
    data = json.load(f)
    for tweet in data:
        all_tweets.append(tweet)
        user_screen_name.append(tweet['user']['screen_name'])
        tweet_id.append(tweet['id'])
        tweet_created_at.append(tweet['created_at'])
        followers_count.append(tweet['user']['followers_count'])
        if 'retweeted_status' in tweet:
            retweet_id.append(tweet['retweeted_status']['id'])
            retweet_created_at[tweet['id']] = tweet['created_at']
            tweet_text[tweet['retweeted_status']['id']] = tweet['retweeted_status']['text']
            if tweet['retweeted_status']['id'] == 1040712546808983555:
                print("RETWEET!!" + str(1040712546808983555))
                print(tweet['retweeted_status']['text'])
        else:
            retweet_id.append(0)
            tweet_text[tweet['id']] = tweet['text']
        inc += 1
        val_inc += 1
        if val_inc > 10000:
            val = val + 10000
            print(str(val))
            val_inc = 0
f.close()

print("End:" + str(datetime.datetime.now()))
```

```python
import collections

retweet_id_counter=collections.Counter(retweet_id)

top1000_retweets = retweet_id_counter.most_common(1000)
top500_retweets = retweet_id_counter.most_common(500)
top36_retweets = retweet_id_counter.most_common(36)
top20_retweets = retweet_id_counter.most_common(20)

top20_tweet_color_indices = {}
top20_most_common_tweet_ids = []
top36_most_common_tweet_ids = []
top500_most_common_tweet_ids = []
top1000_most_common_tweet_ids = []
for i in range(len(top500_retweets)):
    top500_most_common_tweet_ids.append(int(top500_retweets[i][0]))
for i in range(len(top36_retweets)):
    top36_most_common_tweet_ids.append(int(top36_retweets[i][0]))
for i in range(len(top20_retweets)):
    top20_tweet_color_indices[top20_retweets[i][0]] = i
    top20_most_common_tweet_ids.append(top20_retweets[i][0])
for i in range(len(top1000_retweets)):
    top1000_most_common_tweet_ids.append(top1000_retweets[i][0])
```

```python
days_per_month = {'Jan':31,'Feb':28,'Mar':31,'Apr':30,'May':31,'Jun':30,'Jul':31,'Aug':31,'Sep':30,'Oct':31,'Nov':30,'Dec':31}
days_per_month_cumulative = {'Jan':0,'Feb':31,'Mar':59,'Apr':90,'May':120,'Jun':151,'Jul':181,'Aug':212,'Sep':243,'Oct':273,'Nov':304,'Dec':334}
print(str(days_per_month['Jan']))
```

```python
outfile = "idlib_top500_retweet_frequency_counts_matrix_1500sec.csv"
fout = open(outfile, 'w')

fc_outfile = "idlib_top500_retweet_frequency_counts_color_index_matrix_1500sec.csv"
fcout = open(fc_outfile, 'w')

for tweet_inc in range(1,len(top500_most_common_tweet_ids)):

    tweet = top500_most_common_tweet_ids[tweet_inc]
    print(str(tweet_inc))
    total_time = {}
    tweet_frequency_count = []
    followers_count_frequency_count = []

    threshold_epoch = 1500
    threshold = 0
    count_temp = 0
    followers_count_temp = 0

    for i in range(0,len(tweet_id)):

        # Parse the date and time into seconds for total_time and use for translate_y
        date_array = tweet_created_at[i].split(" ")
        time_var = date_array[3]
        time_array = date_array[3].split(":")
        month_seconds = days_per_month_cumulative[date_array[1]] * 86400 # 31 * 24 * 60 * 60
        day_seconds = float(date_array[2]) * 86400 # 24 * 60 * 60
        hour_seconds = float(time_array[0]) * 3600
        minute_seconds = float(time_array[1]) * 60
        seconds = float(time_array[2]) * 1
        if i == 0:
            start_time = float(seconds) + float(minute_seconds) + float(hour_seconds) + day_seconds + month_seconds
            total_time[i] = 0
        else:
            total_time[i] = float(seconds) + float(minute_seconds) + float(hour_seconds) + day_seconds + month_seconds - start_time
        if total_time[i] < threshold:
#             print(str(tweet_inc) + " " + str(i) + " " + str(tweet_id[i] + " " + str(tweet)))
            if int(tweet_id[i]) == int(tweet):
#                 print("Hit")
                count_temp += 1
                followers_count_temp += int(followers_count[i])
            if int(retweet_id[i]) == int(tweet):
#                     print("Hit")
                    count_temp += 1
                    followers_count_temp += int(followers_count[i])
        else:
            threshold += threshold_epoch
            tweet_frequency_count.append(count_temp)
            count_temp = 0
            followers_count_frequency_count.append(followers_count_temp)
            followers_count_temp = 0
    for i in range(len(tweet_frequency_count)):
        if i == 0:
            outputstring = str(tweet_frequency_count[i])
            followers_count_outputstring = str(followers_count_frequency_count[i])
        else:
            outputstring = outputstring + "," + str(tweet_frequency_count[i])
            followers_count_outputstring = followers_count_outputstring + "," + str(followers_count_frequency_count[i])
    outputstring = outputstring + "\n"
    followers_count_outputstring = followers_count_outputstring + "\n"
    fout.write(outputstring)
    fcout.write(followers_count_outputstring)

fout.close()
fcout.close()
```
