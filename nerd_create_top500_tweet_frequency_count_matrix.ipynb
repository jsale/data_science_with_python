```python
inc = 0
val = 0
val_inc = 0
all_tweets = []
tweet_id = []
tweet_created_at = []
followers_count = []
user_screen_name = []
retweet_id = []
retweeted_status = []
language = []
utc_offset = []
retweet_created_at = {}
tweet_text = {}
filename = 'tweet_data/nerd.json'
print("Start:" + str(datetime.datetime.now()))

with open(filename, 'r') as f:
    data = json.load(f)
    for tweet in data:
#         print(str(inc))
        all_tweets.append(tweet)
        user_screen_name.append(tweet['user']['screen_name'])
        tweet_id.append(tweet['id'])
        tweet_created_at.append(tweet['created_at'])
        followers_count.append(tweet['user']['followers_count'])
        if 'retweeted_status' in tweet:
            retweet_id.append(tweet['retweeted_status']['id'])
            retweet_created_at[tweet['id']] = tweet['created_at']
            tweet_text[tweet['id']] = tweet['retweeted_status']['text']
        else:
            retweet_id.append(0)
            tweet_text[tweet['id']] = tweet['text']
        inc += 1
        val_inc += 1
        if val_inc > 10000:
            val = val + 10000
            print(str(val))
            val_inc = 0
f.close()

print("End:" + str(datetime.datetime.now()))
```

```python
outfile = "nerd_top500_retweet_frequency_counts_matrix_120sec.csv"
fout = open(outfile, 'w')

fc_outfile = "nerd_top500_retweet_frequency_counts_color_index_matrix_120sec.csv"
fcout = open(fc_outfile, 'w')

for tweet_inc in range(1,len(top500_most_common_tweet_ids)):

    tweet = top500_most_common_tweet_ids[tweet_inc]
    print(str(tweet_inc))
    total_time = {}
    tweet_frequency_count = []
    followers_count_frequency_count = []

    threshold_epoch = 1500
    threshold = 0
    count_temp = 0
    followers_count_temp = 0

    for i in range(0,len(tweet_id)):

        # Parse the date and time into seconds for total_time and use for translate_y
        date_array = tweet_created_at[i].split(" ")
        time_var = date_array[3]
        time_array = date_array[3].split(":")
        month_seconds = days_per_month_cumulative[date_array[1]] * 86400 # 31 * 24 * 60 * 60
        day_seconds = float(date_array[2]) * 86400 # 24 * 60 * 60
        hour_seconds = float(time_array[0]) * 3600
        minute_seconds = float(time_array[1]) * 60
        seconds = float(time_array[2]) * 1
        if i == 0:
            start_time = float(seconds) + float(minute_seconds) + float(hour_seconds) + day_seconds + month_seconds
            total_time[i] = 0
        else:
            total_time[i] = float(seconds) + float(minute_seconds) + float(hour_seconds) + day_seconds + month_seconds - start_time
        if total_time[i] < threshold:
#             print(str(tweet_inc) + " " + str(i) + " " + str(tweet_id[i] + " " + str(tweet)))
            if int(tweet_id[i]) == int(tweet):
#                 print("Hit")
                count_temp += 1
                followers_count_temp += int(followers_count[i])
            if int(retweet_id[i]) == int(tweet):
#                     print("Hit")
                    count_temp += 1
                    followers_count_temp += int(followers_count[i])
        else:
            threshold += threshold_epoch
            tweet_frequency_count.append(count_temp)
            count_temp = 0
            followers_count_frequency_count.append(followers_count_temp)
            followers_count_temp = 0
    for i in range(len(tweet_frequency_count)):
        if i == 0:
            outputstring = str(tweet_frequency_count[i])
            followers_count_outputstring = str(followers_count_frequency_count[i])
        else:
            outputstring = outputstring + "," + str(tweet_frequency_count[i])
            followers_count_outputstring = followers_count_outputstring + "," + str(followers_count_frequency_count[i])
    outputstring = outputstring + "\n"
    followers_count_outputstring = followers_count_outputstring + "\n"
    fout.write(outputstring)
    fcout.write(followers_count_outputstring)

fout.close()
fcout.close()
```
